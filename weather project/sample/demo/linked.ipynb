```python
import requests

# Function to get weather-related news from NewsAPI
def get_weather_news(place, country_code='IN'):
    news_api_key = '568f66032bf6460e9dcdb60954ee4ade'
    
    # Refined query: Searching news specific to the place (city) and country code, and including weather-related terms (rain, storm, cyclone, flood)
    query = f'{place} AND {country_code} AND (rain OR storm OR cyclone OR flood)'
    
    url = f'https://newsapi.org/v2/everything?q={query}&apiKey={news_api_key}'
    response = requests.get(url)
    news_data = response.json()
    
    # Check if the request was successful
    if news_data['status'] == 'ok':
        # Filter the articles to ensure they are related to the place
        filtered_articles = [article for article in news_data['articles'] if place.lower() in article['title'].lower()]
        return filtered_articles
    else:
        return []

# Example usage
place = 'Chennai'
weather_news = get_weather_news(place)

print(f"Weather-related News Articles for {place} (Rain, Storm, Cyclone, Flood):")

if weather_news:
    for article in weather_news:
        print(f"Title: {article['title']}")
        print(f"Description: {article['description']}")
        print(f"URL: {article['url']}\n")
else:
    print(f"No weather-related news found for {place}.")
```

```python
import requests
import datetime

def get_weather_forecast(lat, lon, api_key):
    url = f"https://api.openweathermap.org/data/2.5/onecall?lat={lat}&lon={lon}&exclude=current,minutely,hourly&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()

    if response.status_code == 200:
        print(f"Weather forecast for the next 7 days:")

        # Iterate through the daily forecast
        for day in data['daily']:
            dt = datetime.datetime.fromtimestamp(day['dt']).strftime('%Y-%m-%d')
            temp = day['temp']['day']
            weather = day['weather'][0]['description']
            print(f"{dt}: {weather}, Temp: {temp}°C")
    else:
        print("Failed to retrieve weather data")

# Example usage
lat = 13.0827  # Chennai latitude
lon = 80.2707  # Chennai longitude
api_key = "b7619db6ddcb6b1e7f3784aeb489d704"  # Replace with your OpenWeatherMap API key

get_weather_forecast(lat, lon, api_key)
```
